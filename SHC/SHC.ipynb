```python
import pandas as pd
```

```python
input=pd.read_csv('input.csv',header=None)
```

```python
input.columns=['Temperature', 'Moisture', 'Nitrogen', 'Phosphorous', 'Potassium',
       'pH', 'Electric Conductivity']
input
```

Conversion of N, P, K EC units

```python
for i in range(len(input)):
    a=1.35*10*(input['Nitrogen'][i])
    input['Nitrogen'][i]=a
    b=1.35*10*(input['Phosphorous'][i])
    input['Phosphorous'][i]=b
    c=1.35*10*(input['Potassium'][i])
    input['Potassium'][i]=c
    d=0.1*(input['Electric Conductivity'][i])
    input['Electric Conductivity'][i]=d
```

```python
input=input[['Nitrogen', 'Phosphorous', 'Potassium', 'pH',
       'Electric Conductivity', 'Temperature', 'Moisture']]
```

```python
parameters=input.columns
inputs=input.iloc[0:1,:]
inputs
```

```python
SHC=input.melt( 
        var_name="Parameter", 
        value_name="Test value")
SHC
```

Tempertaure=> 20-30
Moisture=> (Paddy, Sugarcane=> 80-85), (Cotton, Maize=> 50-60)=> 50-75%
N=> 280-560
P=> 11-26
K=> 120-280
pH=> >7, =7, <7
Electrical Conductivity=> 0-2 dS/m

```python
std=['280-560','11-26','120-280','7, Neutral','0-2','20-30','50-75']
```

```python
unit=['Kg/Ha','Kg/Ha','Kg/Ha','H Potenz','dS/m','°C','%']
```

```python
SHC['Unit']=unit
```

```python
SHC
```

```python
SHC['Rating']=''
```

```python
# N
if (SHC['Test value'][0]<280):
    SHC['Rating'][0]='Low'
elif(SHC['Test value'][0]>=280 and SHC['Test value'][0]<=560):
    SHC['Rating'][0]='Medium'
elif(SHC['Test value'][0]>560):
    SHC['Rating'][0]='High'
# P
if (SHC['Test value'][1]<11):
    SHC['Rating'][1]='Low'
elif(SHC['Test value'][1]>=11 and SHC['Test value'][1]<=26):
    SHC['Rating'][1]='Medium'
elif(SHC['Test value'][1]>26):
    SHC['Rating'][1]='High'
# K
if (SHC['Test value'][2]<120):
    SHC['Rating'][2]='Low'
elif(SHC['Test value'][2]>=120 and SHC['Test value'][2]<=280):
    SHC['Rating'][2]='Medium'
elif(SHC['Test value'][2]>280):
    SHC['Rating'][2]='High'
# pH
if (SHC['Test value'][3]<6):
    SHC['Rating'][3]='Acidic'
elif(SHC['Test value'][3]>=6 and SHC['Test value'][3]<7):
    SHC['Rating'][3]='Slightly Acidic'
elif(SHC['Test value'][3]==7):
    SHC['Rating'][3]='Neutral'
elif(SHC['Test value'][3]>7 and SHC['Test value'][3]<=8):
    SHC['Rating'][3]='Slightly Basic'
elif(SHC['Test value'][3]>8):
    SHC['Rating'][3]='Basic'
# EC
if (SHC['Test value'][4]<0):
    SHC['Rating'][4]='Low'
elif(SHC['Test value'][4]>=0 and SHC['Test value'][4]<=2):
    SHC['Rating'][4]='Normal'
elif(SHC['Test value'][4]>2):
    SHC['Rating'][4]='High'
# Temperature
if (SHC['Test value'][5]<20):
    SHC['Rating'][5]='Cool'
elif(SHC['Test value'][5]>=20 and SHC['Test value'][5]<=30):
    SHC['Rating'][5]='Ideal'
elif(SHC['Test value'][5]>30):
    SHC['Rating'][5]='Hot'
# Moisture
if (SHC['Test value'][6]<50):
    SHC['Rating'][6]='Low'
elif(SHC['Test value'][6]>=50 and SHC['Test value'][6]<=75):
    SHC['Rating'][6]='Medium'
elif(SHC['Test value'][6]>75):
    SHC['Rating'][6]='High'
```

```python
SHC['Normal Level']=std
```

```python
print("The soil health card for the given Parameters is: ")
print(SHC)
```

Colouring the SHC

```python
def color_rating(val):
    if val == 'Low':
        color = 'orange'
    elif val=='High':
        color = 'Red'
    elif val=='Normal':
        color='Grey'

    elif val == 'Ideal':
        color = 'green'
    elif val=='Hot':
        color='Red'
    elif val=='Cool':
        color='SkyBlue'

    elif val == 'Neutral':
        color = 'Grey'
    elif val== 'Acidic' or val=='Slightly Acidic':
        color='Red'
    elif val== 'Basic' or val=='Slightly Basic':
        color='Orange'
    
    return 'background-color: %s' % color

SHC_c = SHC.style.applymap(color_rating, subset=['Rating'])

# Coloured pdf
print("Soil Health Card")
SHC_c
```

```python
import pandas as pd
from io import StringIO, BytesIO
from xhtml2pdf import pisa
from string import Template as HTMLTemplate
```

```python
# Convert dataframes to HTML
html1 = SSD1_farmer1.to_html()
html2 = SSD2_farmer1.to_html()
html3 = SHC_c.to_html()

# Create HTML template
html_template = HTMLTemplate('''
   <html>
     <head>
       <style>
         table, th, td {
           border: 1px solid black;
           border-collapse: collapse;
           padding: 5px;
         }
       </style>
     </head>
     <body>
       <h1>Farmer Details</h1>
       $html1
       <h1>Farm Details</h1>
       $html2
       <h1>Soil Health Card</h1>
       $html3
     </body>
   </html>
''')
```

```python
# Merge HTML content
html = html_template.substitute(html1=html1, html2=html2, html3=html3)
```

```python
# Convert HTML to PDF
pdf = BytesIO()
pisa.CreatePDF(BytesIO(html.encode('utf-8')), pdf)

# Save PDF to file
with open('SoilHealthCard.pdf', 'wb') as f:
    f.write(pdf.getvalue())
```

```python
import pandas as pd
from jinja2 import Template

# Create the DataFrame
df = pd.DataFrame({
    'Parameter': ['Nitrogen', 'Phosphorous', 'Potassium', 'pH', 'Electric Conductivity', 'Salinity', 'Temperature', 'Moisture'],
    'Test value': [27.0, 40.5, 94.5, 7.0, 3.6, 3.6, 26.2, 2.1],
    'Unit': ['Kg/Ha', 'Kg/Ha', 'Kg/Ha', 'H Potenz', 'dS/m', 'dS/m', '°C', '%'],
    'Rating': ['Low', 'High', 'Low', 'Neutral', 'High', 'Normal', 'Ideal', 'Low'],
    'Normal Level': ['280-560', '11-26', '120-280', '7, Neutral', '0-2', '2-4', '20-30', '50-75']
})

# Define the Jinja2 template
template = Template("""
<html>
<head>
  <style>
    table, th, td {
      border: 1px solid black;
      border-collapse: collapse;
      padding: 5px;
    }
    th.Parameter {
      background-color: #99CCFF;
    }
    th.Rating {
      background-color: #FFCC99;
    }
    td.Low {
      background-color: #FFCCCC;
    }
    td.High {
      background-color: #CCFFCC;
    }
    td.Normal {
      background-color: #FFFFCC;
    }
    td.Ideal {
      background-color: #CCFFFF;
    }
  </style>
</head>
<body>
<table>
  <thead>
    <tr>
      <th class="Parameter">Parameter</th>
      <th>Test value</th>
      <th>Unit</th>
      <th class="Rating">Rating</th>
      <th>Normal Level</th>
    </tr>
  </thead>
  <tbody>
    {% for index, row in df.iterrows() %}
      <tr>
        <td class="Parameter">{{ row['Parameter'] }}</td>
        <td>{{ row['Test value'] }}</td>
        <td>{{ row['Unit'] }}</td>
        <td class="{{ row['Rating'] }}">{{ row['Rating'] }}</td>
        <td>{{ row['Normal Level'] }}</td>
      </tr>
    {% endfor %}
  </tbody>
</table>
</body>
</html>
""")

# Render the template with the DataFrame data
html = template.render(df=df)

# Write the HTML file
with open('output.html', 'w') as f:
    f.write(html)
```
